## initial setting

```python
import numpy as np
import matplotlib.pyplot as plt
import os.path
%matplotlib inline
```

```python
import rhk.rhksm4 as rs4
```

```python
sm4=rs4.SM4File("data3293FFT.sm4")
```

```python
[child.objname for child in  sm4.children    ]
```

```python
pageheader = sm4.children[0].children[0].pages[0].children[0]
page = sm4.children[0].children[0].pages[0].children[1]
```

```python
pageheader.x_scale, pageheader.y_scale, pageheader.z_scale, pageheader.x_size, pageheader.y_size, pageheader.z_offset
```

```python
sm4data = np.array(page.data).reshape(pageheader.x_size, pageheader.y_size)
sm4data = sm4data * pageheader.z_scale + pageheader.z_offset
plt.imshow(sm4data)
```
